```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
spamdata = np.genfromtxt("spambase.data",delimiter=",")
X, Y = spamdata[:,0:57], spamdata[:,-1]
```

```python
pca = PCA(n_components=2)
Xpca = pca.fit_transform(X)
```

```python
colors = ['r','g','b']

plt.rcParams['figure.figsize'] = (10,7)

mask=np.where(Y==0)[0] 
plt.scatter(Xpca[mask,0],Xpca[mask,1],s=1,c=colors[0],alpha=0.75,label="Non-Spam")

mask=np.where(Y==1)[0] 
plt.scatter(Xpca[mask,0],Xpca[mask,1],s=1,c=colors[1],alpha=0.75,label="Spam")

plt.legend()
plt.show()
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
Xpca = pca.fit_transform(X)

plt.rcParams['figure.figsize'] = (10,7)

mask=np.where(Y==0)[0] 
plt.scatter(Xpca[mask,0],Xpca[mask,1],s=1,c=colors[0],alpha=0.75,label="Non-Spam")

mask=np.where(Y==1)[0] 
plt.scatter(Xpca[mask,0],Xpca[mask,1],s=1,c=colors[1],alpha=0.75,label="Spam")

plt.legend()
plt.show()
```
